```python
import pandas as pd
```

### First Step:
importing and reading in the data

```python
df = pd.read_csv(r'C:\Users\Owner\Documents\Coding Temple\week_6\day_3\student\student-mat.csv',sep=';')
df.head()
```

### Second Step
Column Transformation for better readability

```python
df.columns = df.columns.str.lower()
# df.columns.to_list()
column_trans = {
    'famsize':'family_size',
    'pstatus':'parent_cohab_status',
    'medu':'mother_education',
    'fedu':'father_education',
    'mjob':'mother_job',
    'fjob':'father_job',
    'traveltime':'travel_time',
    'studytime':'study_time',
    'schoolsup':'school_support',
    'famsup':'family_support',
    'famrel':'family_relationship',
    'freetime':'free_time',
    'goout':'outside_socialness',
    'dalc':'workday_alcohol_drink',
    'walc':'weekend_alcohol_drink',
    'g1':'first_period_grade',
    'g2':'second_period_grade',
    'g3':'final_grade'
}
df.rename(columns=column_trans, inplace=True)
df.columns.to_list()
```

```python
# writing out the names for better readibility
df.school = df.school.str.replace('GP', 'Gabriel Pereira').str.replace('MS', 'Mousinho da Silveira')
df.sex = df.sex.str.replace('F','Female').str.replace('M','Male')
df.address = df.address.str.replace('U','Urban').str.replace('R','Rural')
df.family_size = df.family_size.str.replace('LE3', '>or= 3').str.replace('GT3','<3')
df.parent_cohab_status = df.parent_cohab_status.str.replace('T','Together').str.replace('A','Apart')
```

reading and transforming the second set of data

```python
df2 = pd.read_csv(r'C:\Users\Owner\Documents\Coding Temple\week_6\day_3\student\student-por.csv',sep=';')
df2.columns = df2.columns.str.lower()
column_trans = {
    'famsize':'family_size',
    'pstatus':'parent_cohab_status',
    'medu':'mother_education',
    'fedu':'father_education',
    'mjob':'mother_job',
    'fjob':'father_job',
    'traveltime':'travel_time',
    'studytime':'study_time',
    'schoolsup':'school_support',
    'famsup':'family_support',
    'famrel':'family_relationship',
    'freetime':'free_time',
    'goout':'outside_socialness',
    'dalc':'workday_alcohol_drink',
    'walc':'weekend_alcohol_drink',
    'g1':'first_period_grade',
    'g2':'second_period_grade',
    'g3':'final_grade'
}
df2.rename(columns=column_trans, inplace=True)
df2.school = df2.school.str.replace('GP', 'Gabriel Pereira').str.replace('MS', 'Mousinho da Silveira')
df2.sex = df2.sex.str.replace('F','Female').str.replace('M','Male')
df2.address = df2.address.str.replace('U','Urban').str.replace('R','Rural')
df2.family_size = df2.family_size.str.replace('LE3', '>or= 3').str.replace('GT3','<3')
df2.parent_cohab_status = df2.parent_cohab_status.str.replace('T','Together').str.replace('A','Apart')
df2.head()
```

Merging both DataFrames to get only the students who are in both classes. According to the student text file, there are 382 similar students in both.

```python
df3 = pd.merge(df, df2, on=("school","sex",
                "age","address","family_size","parent_cohab_status",
                "mother_education","father_education","mother_job",
                "father_job","reason","nursery","internet"), how='inner')
df3.info()
```
